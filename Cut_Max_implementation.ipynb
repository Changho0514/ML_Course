```python
# CUT MIX  
from PIL import Image
import numpy as np


def rand_bbox(size, lam):
    """
      crop   
    size:  
    lam:       
    output: crop   bbx1, bby1, bbx2, bby2
    """
  
    W = len(size[0])
    H = len(size[1])
    cut_rat = np.sqrt(1. - lam) #  
    # print('cut_rat :', cut_rat) # 0.83666

    #  W, H  (int) 
    cut_w = np.int(W * cut_rat)
    cut_h = np.int(H * cut_rat)

    # Uniform   x,y  
    cx = np.random.randint(W)
    cy = np.random.randint(H)

    # numpy clip           , cut size //2   
    bbx1 = np.clip(cx - cut_w // 2, 0, W)
    bby1 = np.clip(cy - cut_h // 2, 0, H)
    bbx2 = np.clip(cx + cut_w // 2, 0, W)
    bby2 = np.clip(cy + cut_h // 2, 0, H)

    return bbx1, bby1, bbx2, bby2

def cut_generator(img1, img2, lam = 0.3):
    """ 
    Cut mix   , img1 img2   
    output:   img1
    """
    bbx1, bby1, bbx2, bby2 = rand_bbox(img2, lam)
    img1[bbx1:bbx2, bby1:bby2,:] = img2[bbx1:bbx2, bby1:bby2,:]
     
    return img1

# Test  
#     jpeg 
dir1 = '/content/drive/MyDrive/dog.jpg'
dir2 = '/content/drive/MyDrive/cat.jpg' 

img1 = np.array(Image.open(dir1).convert('RGB').resize((256,256)))
img2 = np.array(Image.open(dir2).convert('RGB').resize((256,256)))

# print(img1[0])
# print(img2.shape)
img = cut_generator(img1,img2)
Image.fromarray(img)
```
